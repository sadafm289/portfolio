## My Role as a Scrum Master

As a scrum master, it is your job to keep your group on task, coordinate with your group, and be organized. Being a scrum master, I always made sure my team turned in everything on time and kept track of new updates. One main thing that helped make my group be organized is assigning roles. To split work, everyone would do one thing and we would combine it into one. For example, when we taught arrays, everyone spoke about one part in the lesson demonstrating what we individually worked on.

For the murder mystery game, we all combined our ideas onto a whiteboard and chose as a group what we wanted our level to look like. This helped us create our level because we already knew what we were going to do. So in the end, we created our level successfully with no errors.
