## Night at the Museum

At night of the museum, there were multiple electives/classrooms which showcased their creativity and design through projects, items, and presentations. For example, in the art classrooms, they had many tables displaying their creations for others to observe. In the computer science classroom, students from APCSP and APCSA presented their projects on screens. As a CSSE student, it was very interesting getting to see what APCSP and APCSA students are working on and building!
